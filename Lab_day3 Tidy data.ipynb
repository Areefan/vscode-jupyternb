```python
import pandas as pd 
```

```python
data = pd.read_excel("data/DATA_PREP_files/data03.xlsx" , header=[0,1])
data
```

### แก้ไขข้อมูลที่เปน คอลัม 2 แถว

```python
pd.MultiIndex.from_frame(data)
```

```python
data_1999 = pd.DataFrame()
data_1999["country"] = data["Unnamed: 0_level_0","country"]
data_1999['year'] = 1999
data_1999['cases'] = data[(1999,"cases")]
data_1999['population'] = data[(1999, "population")]
data_1999
```

```python
data_2000 = pd.DataFrame()
data_2000["country"] = data["Unnamed: 0_level_0","country"]
data_2000['year'] = 2000
data_2000['cases'] = data[(2000,"cases")]
data_2000['population'] = data[(2000, "population")]
data_2000
```

```python
new_data = pd.concat([data_1999, data_2000]).sort_values(['country', 'year']).reset_index(drop=True)

new_data
```

#### แก้ไขข้อมูล 2 variable

```python
data2 = pd.read_excel("data/DATA_PREP_files/data02.xlsx")
data2
```

```python
data_cases = data2[data2["type"] == "cases"].drop("type", axis=1) 
data_cases = data_cases.rename(columns = { "count" : 'cases'})
data_cases
```

```python
data_population = data2[data2["type"] == "population"].drop("type", axis=1) 
data_population = data_population.rename(columns = { "count" : 'population'})
data_population
```

### merge

```python
data_cases.merge(data_population, on = ["country", "year"])
```
